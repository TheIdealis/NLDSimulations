```python
import pylab as plt
import numpy as np
import subprocess
```

## Parameter

```python
k0 = 1
n = 1
delta = 1
I0 = 1
beta = .1
```

```python
v = np.linspace(-7, 7, 1000)
def pot(k0, n, delta, I0, beta):
    return - n*k0 * (beta * v**2 + delta * k0 * I0 * np.log(1 + v**2/I0))
```

```python
plt.figure(figsize=(17,4))
plt.subplot(141)
plt.plot(v, pot(k0, n, delta, I0, beta))
# plt.yticks([])
plt.title('$\delta > 0$, $\\beta > 0$', fontsize=20)
plt.subplot(142)
plt.plot(v, pot(k0, n, -delta, I0, beta))
# plt.yticks([])
plt.title('$\delta < 0$, $\\beta > 0$', fontsize=20)
plt.subplot(143)
plt.plot(v, pot(k0, n, delta, I0, -beta))
# plt.yticks([])
plt.title('$\delta > 0$, $\\beta < 0$', fontsize=20)
plt.subplot(144)
plt.plot(v, pot(k0, n, -delta, I0, -beta))
# plt.yticks([])
plt.title('$\delta < 0$, $\\beta < 0$', fontsize=20)
plt.savefig('../homework/s3_p1.pdf')
subprocess.call(['pdfcrop', '../homework/s3_p1.pdf', '../homework/s3_p1.pdf'])
```
