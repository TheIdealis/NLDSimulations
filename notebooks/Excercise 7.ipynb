```python
%matplotlib inline

import numpy as np
import pylab as plt
# plt.style.use('default')
import subprocess as sub
```

```python
fig = plt.figure(figsize=(12, 4))
ax = [None]*3

fig.subplots_adjust(left = 0.25, bottom = 0.25)

x = np.linspace(0, 10, 300)

ax[0] = plt.subplot(131)
plt.plot(x, (x - 4)**2 + 5)
plt.ylim([-1, 45])
ax[0].set_ylabel('$S(z)$')

ax[1] = plt.subplot(132)
f = 0.4 * (x - 8)**2 - 2
plt.plot(x[f > 0], f[f > 0])

ax[2] = plt.subplot(133)
f = -0.4 * (x - 2)**2 + 7
plt.plot(x[f > 0], f[f > 0])

for axes in ax:
    axes.set_xticks([0])
    axes.set_yticks([0])
    axes.set_xlim([-0.2, 10.2])
    axes.plot([-1, 11], [0, 0], 'k--')
    axes.set_xlabel('$z$')
    
filename = 'fig_blatt7.pdf'

plt.savefig(filename)
sub.call(['pdfcrop'] + 2 * [filename])
```
